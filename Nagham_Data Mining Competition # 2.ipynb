### Answers to Questions:
🌈- **Why a simple linear regression model (without any activation function) is not good for classification task, compared to Perceptron/Logistic regression?**
  - Linear regression outputs continuous values, making it unsuitable for binary classification where discrete class labels are needed. Perceptron and logistic regression models use activation functions to map the output to a binary decision boundary, making them more appropriate for classification tasks.

🌈- **What's a decision tree and how it is different from a logistic regression model?**
  - A decision tree is a non-parametric supervised learning method used for classification and regression tasks. It creates a tree-like structure where each internal node represents a feature, each branch represents a decision based on that feature, and each leaf node represents the outcome.
  - Logistic regression is a parametric linear model used for binary classification. It models the probability that a given input belongs to a particular class using a logistic function.

🌈- **What's the difference between grid search and random search?**
  - Grid search exhaustively searches through a specified grid of hyperparameters, evaluating all possible combinations. It is more exhaustive but may be computationally expensive.
  - Random search randomly samples hyperparameters from specified distributions. It is less exhaustive but can be more efficient in high-dimensional parameter spaces.

🌈- **What's the difference between Bayesian search and random search?**
  - Bayesian search uses probabilistic models to select the next hyperparameter values to evaluate based on past observations. It adapts and refines its search based on the results of previous evaluations.
  - Random search, as mentioned earlier, randomly samples hyperparameters from specified distributions without considering past evaluations.

```python
# Define preprocessing steps
# Identify numeric features and create a pipeline to impute missing values with the mean and scale numerical features
numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Impute missing values with mean
    ('scaler', StandardScaler())  # Scale numerical features
])
# Identify categorical features and create a pipeline to impute missing values with the most frequent value and one-hot encode categorical features
categorical_features = X_train.select_dtypes(include=['object']).columns
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values with most frequent value
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # One-hot encode categorical features
])

# Combine preprocessing steps for numerical and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

##### This code defines preprocessing pipelines for numerical and categorical features and then combines them using ColumnTransformer.
The code defines the preprocessing steps required before training a machine learning model:

1. **Numeric Features Preprocessing**:
   - Identifies numeric features in the training data (`X_train`) using the `select_dtypes` method.
   - Creates a pipeline (`numeric_transformer`) consisting of two steps:
     - Imputation: Missing values in numeric features are imputed using the mean value of each feature.
     - Scaling: Numeric features are scaled using StandardScaler to standardize their values.

2. **Categorical Features Preprocessing**:
   - Identifies categorical features in the training data using the `select_dtypes` method.
   - Creates a pipeline (`categorical_transformer`) consisting of two steps:
     - Imputation: Missing values in categorical features are imputed using the most frequent value of each feature.
     - One-Hot Encoding: Categorical features are one-hot encoded to convert them into numerical format, with a separate binary column created for each category.

3. **Combining Preprocessing Steps**:
   - The `ColumnTransformer` combines the numeric and categorical preprocessing steps into a single transformer, specifying separate transformations for numeric and categorical features.
   - Two transformer tuples are provided: one for numeric features (`('num', numeric_transformer, numeric_features)`) and one for categorical features (`('cat', categorical_transformer, categorical_features)`).
   - The resulting `preprocessor` object can be used in a machine learning pipeline to preprocess both numeric and categorical features before model training.

### Trial 16: Support Vector Classifier (SVC) with Bayesian Optimization

In Trial 16, we employ the Support Vector Classifier (SVC) with Bayesian optimization to search for optimal hyperparameters. The objective is to leverage Bayesian optimization to efficiently explore the hyperparameter space and identify configurations that maximize the ROC AUC score.

#### Methodology:

- We construct an SVC pipeline comprising preprocessing steps and the SVC classifier.
- Bayesian optimization is utilized to search for optimal hyperparameters for the SVC model.
- The search space includes hyperparameters such as C (regularization parameter), gamma (kernel coefficient), degree (degree of the polynomial kernel), and kernel type.
- Bayesian optimization iteratively selects hyperparameter configurations based on their expected improvement in the objective function, which in this case is the ROC AUC score.
- We perform Bayesian optimization for a specified number of iterations, considering cross-validation with 2 folds and evaluating the performance using the ROC AUC score.

#### Implementation:

- The SVC pipeline is defined with preprocessing steps and the SVC classifier.
- Search spaces are specified for the hyperparameters, including C, gamma, degree, and kernel type.
- Bayesian optimization is conducted using BayesSearchCV, which iteratively explores the search space to find optimal hyperparameters.
- We set the number of iterations for Bayesian optimization, random state, verbosity level, cross-validation folds, and scoring metric.

#### Plan for Evaluation:

- After Bayesian optimization, we evaluate the performance of the SVC model with the optimized hyperparameters on the validation set.
- We assess the ROC AUC score to determine the effectiveness of the hyperparameter tuning process.
- Comparison with previous trials, particularly those employing different optimization techniques or models

```python
from sklearn.svm import SVC
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

# Define the SVC pipeline
SVC_pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('my_svc', SVC(class_weight='balanced'))
    ]
)

# Define the search space for hyperparameters
search_spaces = {
    'my_svc__C': Real(1e-6, 1e+6, prior='log-uniform'),
    'my_svc__gamma': Real(1e-6, 1e+1, prior='log-uniform'),
    'my_svc__degree': Integer(1, 8),
    'my_svc__kernel': Categorical(['linear', 'poly', 'rbf']),
}

# Perform Bayesian optimization
bayes_search = BayesSearchCV(
    SVC_pipeline,
    search_spaces=search_spaces,  # Provide search_spaces with the parameter distributions
    n_iter=3,
    random_state=0,
    verbose=1,
    cv=2,
    scoring='roc_auc'
)
```

```python
bayes_search.fit(X_train, y_train)
```

```python
print('best score {}'.format(bayes_search.best_score_))
print('best score {}'.format(bayes_search.best_params_))
```

**Observation of Trial 16 - SVC with Bayesian Optimization:**

- The Bayesian optimization process achieved a best score of 0.9575 on the validation set.
- The hyperparameters corresponding to the best score are as follows:
  - C (regularization parameter): 2.3527
  - Degree (degree of the polynomial kernel): 6
  - Gamma (kernel coefficient): 0.0229
  - Kernel type: Polynomial ('poly')
- Compared to previous trials, the Bayesian optimization approach yielded a relatively high ROC AUC score, indicating effective tuning of the SVC model's hyperparameters.
- The selection of polynomial kernel ('poly') suggests that the dataset may exhibit complex nonlinear relationships that are effectively captured by the polynomial kernel.
